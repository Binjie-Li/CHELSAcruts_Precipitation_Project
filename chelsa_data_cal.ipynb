# [Chelsa 值计算](#toc1_)

```python
import pandas as pd

def process_station(station_name, year):
    filepath = f"{station_name}_{year}.csv"
    data = pd.read_csv(filepath)
    # check if there is any missing value
    if data.isnull().values.any():
        try: 
            # fill the missing value with the above value
            data = data.fillna(method='ffill')
        except:
            # fill the missing value with the below value
            data = data.fillna(method='bfill')
            
    # Convert the 'date' column to datetime format
    data['date'] = pd.to_datetime(data['date'])
    
    # Define the quarters based on your requirements
    quarters = {
        "spring": (3, 5),
        "summer": (6, 8),
        "autumn": (9, 11),
        "winter": (12, 2)
    }

    print(f"Station_name: {station_name}")
    print(f"Year: {year}")
    # print segment line
    print('-' * 40)
    print('Year PRCP_mean PRCP_sum')
    print(data['value'].mean())
    print(data['value'].sum())
    print('-' * 40)
    print("Season PRCP_mean PRCP_sum")
    
    # Loop through each quarter
    for quarter_name, (start_month, end_month) in quarters.items():
        if quarter_name == 'winter':
            december_data = data[data['date'].dt.month == 12]
            january_february_data = data[(data['date'].dt.month >= 1) & (data['date'].dt.month <= end_month)]
            quarter_data = pd.concat([december_data, january_february_data])
        else:
            quarter_data = data[(data['date'].dt.month >= start_month) & 
                                (data['date'].dt.month <= end_month)]
        PRCP_mean = quarter_data['value'].mean()
        PRCP_sum = quarter_data['value'].sum()
        print(f"{quarter_name} {PRCP_mean:.2f} {PRCP_sum:.2f}")
# process_station('USC00049152',2003)  # 作为一个示例
```
