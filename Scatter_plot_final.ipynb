```python
import pandas as pd
data = pd.read_csv('final data.csv')
data
```

```python
data = data.iloc[:70,1:11]
data
```

```python
data.columns=['Observation_Annual','Observation_Spring','Observation_Summer','Observation_Autumn','Observation_Winter','Chelsa_Annual','Chelsa_Spring','Chelsa_Summer','Chelsa_Autumn','Chelsa_Winter']
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 假设你已经有了一个名为data的DataFrame
actual_data = data.iloc[1:].apply(pd.to_numeric, errors='coerce')

def calculate_metrics(data, col1, col2):
    cc = data[col1].corr(data[col2])
    rmse = np.sqrt(np.mean((data[col1] - data[col2])**2))
    bias = 100 * np.mean((data[col1] - data[col2]) / data[col1])
    return cc, rmse, bias

def plot_scatter(ax, data, col1, col2, title, xlims, ylims, x_ticks, y_ticks):
    ax.scatter(data[col1], data[col2], label='Data Points')
    ax.set_title(title, weight='bold', fontsize=12)
    ax.set_xlabel(col1)
    ax.set_ylabel(col2)
    
    # 绘制y=x线
    ax.plot(xlims, ylims, color='red', label='Y = X')
    
    ax.set_xlim(xlims)
    ax.set_ylim(ylims)
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    
    cc, rmse, bias = calculate_metrics(data, col1, col2)
    ax.annotate(f'CC={cc:.2f}', (0.05, 0.92), xycoords='axes fraction', fontsize=10, color='black')
    ax.annotate(f'RMSE={rmse:.2f} mm', (0.05, 0.87), xycoords='axes fraction', fontsize=10, color='black')
    ax.annotate(f'Bias={bias:.2f}%', (0.05, 0.82), xycoords='axes fraction', fontsize=10, color='black')
    
    ax.legend(loc='lower right')
    ax.grid(True)

columns_to_plot = [
    ('Chelsa_Spring', 'Observation_Spring'),
    ('Chelsa_Summer', 'Observation_Summer'),
    ('Chelsa_Autumn', 'Observation_Autumn'),
    ('Chelsa_Winter', 'Observation_Winter'),
    ('Chelsa_Annual', 'Observation_Annual')
]

# 定义范围和刻度的示例
ranges_and_ticks = [
    ((0, 1200), (0, 1200), range(0, 1201, 300), range(0, 1201, 300)),
    ((0, 900), (0, 900), range(0, 901, 300), range(0, 901, 300)),
    ((0, 1200), (0, 1200), range(0, 1201, 300), range(0, 1201, 300)),
    ((0, 1200), (0, 1200), range(0, 1201, 400), range(0, 1201, 400)),
    ((0, 3200), (0, 3200), range(0, 3201, 800), range(0, 3201, 800))
]

# 使用gridspec布局
fig = plt.figure(figsize=(15, 10))
gs = fig.add_gridspec(2, 3, hspace=0.5)

axs = [
    fig.add_subplot(gs[0, 0]),
    fig.add_subplot(gs[0, 1]),
    fig.add_subplot(gs[0, 2]),
    fig.add_subplot(gs[1, 0:1]),
    fig.add_subplot(gs[1, 1:2]) 
]


# 为每张图设置标题
titles = ["(a) Spring", "(b) Summer", "(c) Autumn", "(d) Winter", "(e) Annual"]

for ax, (col1, col2), title, (xlims, ylims, x_ticks, y_ticks) in zip(axs, columns_to_plot, titles, ranges_and_ticks):
    plot_scatter(ax, actual_data, col1, col2, title=title, xlims=xlims, ylims=ylims, x_ticks=x_ticks, y_ticks=y_ticks)

plt.tight_layout()
plt.show()
```
