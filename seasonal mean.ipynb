```python
import pandas as pd

def process_stations(cityname):
    filepath = f"{cityname}_2_selected_Stations.csv"
    data = pd.read_csv(filepath)
    
    # Convert the 'DATE' column to datetime format
    data['DATE'] = pd.to_datetime(data['DATE'])
    
    # Define the quarters with correct seasonal mappings
    quarters = {
        "spring": (3, 5),
        "summer": (6, 8),
        "autumn": (9, 11),
        "winter": (12, 2)
    }
    
    # Loop through each unique station
    for station in data['STATION'].unique():
        station_data = data[data['STATION'] == station]
        print(f"Station_name: {station}")
        
        # Loop through each year
        for year in range(2003, 2017):
            print(year)
            yearly_data = station_data[station_data['DATE'].dt.year == year]
            
            # Loop through each quarter
            for quarter_name, (start_month, end_month) in quarters.items():
                if quarter_name == 'winter':
                    december_data = yearly_data[yearly_data['DATE'].dt.month == 12]
                    january_february_data = yearly_data[(yearly_data['DATE'].dt.month >= 1) & (yearly_data['DATE'].dt.month <= end_month)]
                    quarter_data = pd.concat([december_data, january_february_data])
                else:
                    quarter_data = yearly_data[(yearly_data['DATE'].dt.month >= start_month) & (yearly_data['DATE'].dt.month <= end_month)]
                PRCP_mean = quarter_data['PRCP'].mean()
                PRCP_sum = quarter_data['PRCP'].sum()
                print(f"{quarter_name} {PRCP_mean:.2f} {PRCP_sum:.2f}")
        print("\n")
```
